## Data Cleaning

- Our goal 'cleaned' data will be all the dataset can be represented as feature layers in Arcgis. Then we need to convert following dataset into shapefile. 'housing.csv' only have corresponding longitude, latitude info, so we need to use spatial join with U.S ZIP code boundary layer, making it under corresponding ZIP code so that we can perform analysis within ZIP code polygons. Since the U.S ZIP code layer has EPSG 3857 coordinate system, we need to convert the layer to ESPG 4326 in order to make the projection consistent. Also, the spatial join of Arcgis will raise topologic error when we perform it, and we did not find solution upon Arcgis python API within those error. Alternatively, we used geopandas to perform data cleaning and perprocessing, whcih makes the cleaning process much smoother.
```python
# Some of analysis part code is also being put here
# give geometry to the data

housing = gpd.read_file('housing.csv')
housing['geometry'] = housing.apply(lambda x: Point(float(x['longitude']),float(x['latitude'])),axis = 1)
housing = housing.set_geometry('geometry')
housing = gpd.GeoDataFrame(housing, crs={'init':'epsg:4326'}, geometry='geometry')

# we get the zip code data from arcgis online and save it as shp file in order
# to save time

# df = housing.to_crs({'init':'epsg:3857'})
# ca_ly = gis.content.get("947d9a461e6a4f9db25e42c5938ea47c")
# ca_df = pd.DataFrame.spatial.from_layer(ca_ly.layers[0])
# df['longitude'] = df.geometry.x
# df['latitude'] = df.geometry.y
# df_1 = df.spatial.from_xy(df,x_column = 'longitude', y_column = 'latitude', sr = 102100)
# df_1.SHAPE[0].values()

# chaneg crs of the zip code data

cazip = gpd.read_file('cazip.shp')
cazip = gpd.GeoDataFrame(cazip, crs={'init':'epsg:3857'}, geometry='geometry')
cazip = cazip.to_crs({'init':'epsg:4326'})
cazip.to_file('cazip_coordinated.shp')
```

```python
# spatial join the zipcode and the housing
merged = gpd.sjoin(cazip, housing, how='inner', op='contains', lsuffix='left', rsuffix='right')

# merage the data with the same zip code, average the medians
def group_zip(x):
    
    return pd.DataFrame({
        'housing_median_age': x.housing_median_age.astype(float).mean(),
        'total_rooms': x.total_rooms.astype(float).sum(),
        'total_bedrooms': pd.to_numeric(x.total_bedrooms, errors='coerce').sum(),
        'population': x.population.astype(float).sum(),
        'households': x.households.astype(float).sum(),
        'median_income': x.median_income.astype(float).mean(),
        'median_house_value': x.median_house_value.astype(float).mean(),
        'ocean_proximity': [x.ocean_proximity.mode()[0] if len(x.ocean_proximity.mode()) > 0 else x.ocean_proximity.mode()]
    })

# change the geometry for plotingg
housing_grouped = merged.groupby('ZIP_CODE').apply(lambda x: group_zip(x))
housing_grouped = housing_grouped.reset_index().drop('level_1', axis = 1)
housing_grouped['ZIP_CODE'] = housing_grouped.astype(str)
housing_grouped['geometry'] = housing_grouped.ZIP_CODE.apply(lambda x: merged[merged.ZIP_CODE == x].geometry.iloc[0])
housing_grouped = gpd.GeoDataFrame(housing_grouped, crs={'init':'epsg:4326'}, geometry='geometry')
housing_grouped.to_file('housing_grouped.shp')

# compute the value of price per room and price per bedroom in each zip code
housing_grouped = gpd.read_file('housing_grouped.shp')
housing_grouped['price/room'] = housing_grouped\
.apply(lambda x: x.median_hou / (x.total_room / x.households), axis = 1)
housing_grouped['price/bed'] = housing_grouped\
.apply(lambda x: x.median_hou / (x.total_bedr / x.households), axis = 1)
housing_grouped.to_file('housing_grouped.shp')
```

## Summary of products and results

In order to get statistical evidence of which data have the most influence on the housing price. We do a Random Forest Regressor to validate our analysis. We put all the related variables in the original dataset together with the education and hospital information into the random forest regressor and then compute the feature importance from it.

```python
# manually one hot encoding ocean proximity
def transform_ocean(x):
    if x == 'INLAND' or x == 'ISLAND':
        return 0
    elif x == 'NEAR OCEAN':
        return 1
    elif x == '<1H OCEAN':
        return 2
    elif x == 'NEAR BAY':
        return 3
df_e.ocean_prox = df_e['ocean_prox'].apply(transform_ocean)

col = ["housing_me", 'total_room', "total_bedr", "population",
       'households', "median_inc", "ocean_prox", "Bach_Ratio", "has_hospit"]
final = df_e[col]
```

```python
# split our data and choose only the related features
X = final
y_median = df_e["median_hou"]
X_train, X_test, y_train, y_test = train_test_split(X, y_median, test_size=0.25)

# using random forest regressor to predict median housing price
rf = RandomForestRegressor()
parameters = {"min_samples_split": [2,4,6,8,12,16,20]}
clf = GridSearchCV(rf, parameters, return_train_score=True, cv=5)
clf.fit(X_train, y_train)
```

```python
# r2 score and feature importance
print("R2 SCORE = "+str(clf.score(X_test,y_test)))
feature_impo = sorted(dict(zip(final.columns,clf.best_estimator_.feature_importances_)).items(), 
                      key=operator.itemgetter(1), reverse=True)
feature_impo
```

### Prediction for California housing price for the next three years
We used Zillow Home Value Index as our model for prediction the housing price over the next three years. We used `Gradient Boosting Regressor` as our algorithm. We first used spatial join to join first ten features of the original data to the Zillow dataset. And we fed in the ten years data into the model the predict the next month's price. Before that, We used this method to cross validate the price we already have. We the prediction accuracy have approximately 99% rmse in each step. So after 36 steps (3 years), our prediction may have approximately 70% accuracy.

```python
# load zillow data
zillow_sqft = pd.read_csv('Zip_MedianListingPricePerSqft_AllHomes.csv', encoding='latin-1')

# Preprosess zillow's data and join it to our dataset with corresponding zip codes
zillow_sqft['RegionName'] = zillow_sqft['RegionName'].apply(lambda x: str(x))
zillow_sqft['RegionName'] = zillow_sqft['RegionName'].str.zfill(5)
ca_zip = gpd.read_file('cazip_coordinated.shp')
additional = ca_zip.merge(zillow_sqft, left_on = 'ZIP_CODE', right_on = 'RegionName')

# set the geometry of the data
time_series_price = additional
time_series_price = time_series_price.set_index('ZIP_CODE')
time_series_price = time_series_price.T
time_series_price = time_series_price.reset_index()
time_series_price.columns =  ['time'] + [i for i in time_series_price.columns[1:]]
time_series_price.to_csv('time_sireis_price.csv')
time_series_price = pd.read_csv('time_sireis_price.csv', index_col = 0)
time_series = time_series_price.T
time_series = time_series.reset_index()
time_series = ca_zip.merge(time_series, left_on = 'ZIP_CODE',  right_on= 'index')
housing_grouped = gpd.read_file('housing_grouped.shp')
prediction_csv = housing_grouped.merge(time_series, left_on = 'ZIP_CODE', right_on = 'ZIP_CODE')
zips = prediction_csv['ZIP_CODE']
geometry = prediction_csv['geometry_x']

### preprocess our data for prediction###
prediction_csv = prediction_csv\
            .drop(['ZIP_CODE','price/room', 'price/bed',\
                  'geometry_x', 'FID', 'OBJECTID', 'PO_NAME', 'STATE', 'Shape__Are', 'Shape__Len', 
                 'Shape__Len', 'geometry_y', 'index'], axis = 1)

def transform_ocean(x):
    if x == 'INLAND' or x == 'ISLAND':
        return 0
    elif x == 'NEAR OCEAN':
        return 1
    elif x == '<1H OCEAN':
        return 2
    elif x == 'NEAR BAY':
        return 3
    
prediction_csv['ocean_prox'] = prediction_csv['ocean_prox'].apply(transform_ocean)
prediction_csv = prediction_csv.drop([i for i in range(17)], axis = 1)
prediction_csv = prediction_csv.astype(float)
prediction_csv.columns = [i for i in prediction_csv.columns][:10] + [i for i in range(len(prediction_csv.columns))][:-10]
prediction_csv_o = prediction_csv.copy()
prediction_csv = prediction_csv.apply(lambda x: x.fillna(x.mean()))
X = prediction_csv.iloc[:,:-1] # feature data
y = prediction_csv.iloc[:,-1] # target data
ind = [i for i in range(X.shape[1] + 1)]
ind.remove(11)
ix = 111
pred = pd.DataFrame()
for itera in range(36): # set up prediction for 3 months
    print(itera, end = ' ')
    clf = GradientBoostingRegressor()
    clf.fit(X, y)
    X[ix] = y
    X = X.iloc[:, ind]
    y = clf.predict(X)
    ix += 1
    pred[ix] = y
```

```python
# appending zip and geometry and save as a layer
prediction = prediction_csv.T.append(pred.T).T
prediction['zip_code'] = zips
prediction['geometry'] = geometry
layer_ = gpd.GeoDataFrame(prediction)
layer_ = layer_.set_geometry('geometry')
layer_ = layer_.astype('O')
layer_.columns = pd.Series(layer_.columns).astype(str)
layer_.to_file('predictionly.shp')
```

1. What we found in the random forest regressor is that the median income and the ocean proximity have great influence in determining the housing price. With high income and closer ocean proximity, housing tend to have a higher price, which is indeed what in the Homeguru and "The Journal of Real Estate Finance and Economics" said. However, we found something new. The bachelor population ratio also has a great deal influence in determining the housing price, as working age population have more saving and power to pursue mortgages to purchase housing. In contrast, the hospital within 10 mile of housing does not contribute much the housing price. We believe this is due to the fact that lots of California families own cars, which is convenient for them to go to hospital that is outside the ten-mile zone of any hospital. And also, not a lot of people are looking for hospitals as a factor when searching for available housing. Therefore, median income, ocean proximity and bachelor population ratio are the key factors in determining housing price.


2. One of the trade-offs we made is that buffer area of the hospital. It is quite hard to define the buffer zone. (We deleted the natural disasters feature mainly because of this). In the final stage, we choose 10 miles as the buffer distance, since we believe that the 10-mile walking time is quite a lot if there is an emergency issue. We cannot estimate how many people have cars in each zip code area, therefore, walking time becomes our primary concern.  Another concern we make is whether to use high school population or to use bachelor population. We use the bachelor population at last mainly because there are quite a lot of universities in the whole California zone, thus it would be better to use the bachelor population in the feature in order to fully reflect the population education level in the area.
